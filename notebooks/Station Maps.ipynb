```python
station = pd.read_csv("../../mast30034-project-1-janggani/data/raw/Stations - NYC.csv")
```

```python
# specify coordinate system
crs = {'init': 'epsg:4326'}
```

```python
# create points from latitude longitude data
geometry = [Point(xy) for xy in zip(station["Longitude"], station["Latitude"])]

geo_df = gpd.GeoDataFrame(station, crs=crs, geometry = geometry)
```
